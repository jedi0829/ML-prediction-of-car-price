# Data Cleaning

```python
import numpy as np
cars_features = cars_features.replace('?', np.nan)
cars_features
```

```python
cars_features.info()
```

```python
convert_dict = {'one': 1, 
               'two': 2, 
               'three': 3,
               'four': 4,
               'five': 5, 
               'six': 6,
               'seven': 7,
               'eight': 8, 
               'nine': 9,
               'ten': 10, 
                'eleven': 11, 
                'twelve': 12,
               np.nan: np.nan}

def word_to_num(word):
    return convert_dict[word]

cars_features['num-of-doors'] = cars_features['num-of-doors'].apply(word_to_num)
cars_features['num-of-doors']

#cars_features.astype('float')
#cars_features
```

```python
cars_features['num-of-cylinders'] = cars_features['num-of-cylinders'].apply(word_to_num)
cars_features['num-of-cylinders']
```

```python
cars_features = cars_features.astype('float')
cars_features.isnull().sum()
```

```python
# Remove the rows with missing prices
cars_features = cars_features.dropna(subset=['price'])
cars_features.info()
```

```python
#Replace NaN value in the column with the mean of the column
cars_features['normalized-losses'] = cars_features['normalized-losses'].fillna(cars_features['normalized-losses'].mean())
cars_features['normalized-losses'].isnull().sum()
```

```python
cars_features['num-of-doors'] = cars_features['num-of-doors'].fillna(cars_features['num-of-doors'].mean())
cars_features['horsepower'] = cars_features['horsepower'].fillna(cars_features['horsepower'].mean())
cars_features['peak-rpm'] = cars_features['peak-rpm'].fillna(cars_features['peak-rpm'].mean())
cars_features['bore'] = cars_features['bore'].fillna(cars_features['bore'].mean())
cars_features['stroke'] = cars_features['stroke'].fillna(cars_features['stroke'].mean())
```

```python
cars_features.isnull().sum()
```

```python
cars_features.head()
```

Now all the data has been cleaned, with no missing values in any columns. We will now normalize or rescale all the data into the range of 0 to 1, except for 'price' column. 

In order to get the range of 0 to 1, we need to use a "min-max" normalization. 

```python
#Normalize all features except for 'Price', the target column
price_column = cars_features['price']
cars_features = (cars_features - cars_features.min()) / (cars_features.max() - cars_features.min())
cars_features['price'] = price_column
cars_features.head()
```

# Univariate Model

Let's start with some univariate k-nearest neighbors models

```python
#Set up a function to take in parameters including train_column, target_column, df
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error
def knn_train_test(train_column, target_column, df):
    train_set = df.iloc[0: int(df.shape[0] / 2)]
    test_set = df.iloc[int(df.shape[0] / 2):]
    knn = KNeighborsRegressor()
    knn.fit(train_set[[train_column]], train_set[target_column])
    prediction = knn.predict(test_set[[train_column]])
    rmse = (mean_squared_error(test_set[target_column], prediction)) ** (0.5)
    return rmse
```

```python
#put all the features in a list
cols_to_train = cars_features.columns.tolist()
cols_to_train.remove('price')
cols_to_train
```

```python
#loop through the list of features, and collect different rmse for each feature to train
rmse_dict = {}
for col in cols_to_train:
    rmse_dict[col] = knn_train_test(col, 'price', cars_features)
rmse_dict = pd.Series(rmse_dict)
rmse_dict.sort_values()
```

According to the result, using highway-mpg, horsepower, and engine-size columns in a univariate model will generate lowest rmse with default k value. 

Now, we modify the knn_train_test() function to accept different k-values as parameter within (1,3,5,7,9) range for each training feature. 

```python
#modify the function so it can loop through different K values and find rmse for each k value for each feature
def knn_train_test(train_column, target_column, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    rand_df = df.reindex(shuffled_index)
    train_set = rand_df.iloc[0: int(len(rand_df) / 2)]
    test_set = rand_df.iloc[int(len(rand_df) / 2):]
    k_values = [1, 3, 5, 7, 9]
    rmse_by_k = {}
    for k in k_values:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_set[[train_column]], train_set[target_column])
        prediction = knn.predict(test_set[[train_column]])
        rmse = (mean_squared_error(test_set[target_column], prediction)) ** (0.5)
        rmse_by_k[k] = rmse
    return rmse_by_k
```

```python
rmse_all = {}
for col in cols_to_train:
    rmse = knn_train_test(col, 'price', cars_features)
    rmse_all[col] = rmse
rmse_all
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
for k,v in rmse_all.items():
    x = list(v.keys())
    y = list(v.values())
    plt.plot(x, y)
    
```

# Multivariate Model

Now make changes to knn_train_test() function to work with multiple columns

```python
# First, find out which features are the best features (smallest rsme)
feature_avg_rsme = {}
for k,v in rmse_all.items():
    avg_rmse = np.mean(list(v.values()))
    feature_avg_rsme[k] = avg_rmse
pd.Series(feature_avg_rsme).sort_values()
```

```python
#modify the training function to take in not one feature, but a list of multiple features at the same time to train
def knn_train_test(train_columns, target_column, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    train_set = shuffled_df.iloc[0: int(df.shape[0] / 2)]
    test_set = shuffled_df.iloc[int(df.shape[0] / 2):]
    k_list = [5]
    rmse_by_k = {}
    for k in k_list:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_set[train_columns], train_set[target_column])
        prediction = knn.predict(test_set[train_columns])
        rmse = (mean_squared_error(test_set[target_column], prediction)) ** (0.5)
        rmse_by_k[k] = rmse
    return rmse_by_k
```

```python
#run the training model on best two, three, four, and five features separately and collect rmse for each training
results_by_num_best_features = {}
best_2_features = ['engine-size', 'curb-weight']
results_by_num_best_features['best_2_features'] = knn_train_test(best_2_features, 'price', cars_features)

best_3_features = ['engine-size', 'curb-weight', 'city-mpg']
results_by_num_best_features['best_3_features'] = knn_train_test(best_3_features, 'price', cars_features)

best_4_features = ['engine-size', 'curb-weight', 'city-mpg', 'highway-mpg']
results_by_num_best_features['best_4_features'] = knn_train_test(best_4_features, 'price', cars_features)

best_5_features = ['engine-size', 'curb-weight', 'city-mpg', 'highway-mpg', 'width']
results_by_num_best_features['best_5_features'] = knn_train_test(best_5_features, 'price', cars_features)

results_by_num_best_features
```

# Hyperparamter Tuning

Now we need to tune the hyperparameter, the k-value from 1 to 25, and plot the resulting rmse values, and find out which k-value produced the best predicting result. 

```python
#modify the training function to change the k-value list to include the range of 1 to 25
def knn_train_test(train_columns, target_column, df):
    np.random.seed(1)
    shuffled_index = np.random.permutation(df.index)
    shuffled_df = df.reindex(shuffled_index)
    train_set = shuffled_df.iloc[0: int(df.shape[0] / 2)]
    test_set = shuffled_df.iloc[int(df.shape[0] / 2):]
    k_list = list(range(1,26,1))
    rmse_by_k = {}
    for k in k_list:
        knn = KNeighborsRegressor(n_neighbors=k)
        knn.fit(train_set[train_columns], train_set[target_column])
        prediction = knn.predict(test_set[train_columns])
        rmse = (mean_squared_error(test_set[target_column], prediction)) ** (0.5)
        rmse_by_k[k] = rmse
    return rmse_by_k
```

```python
#Choose the best three models to run the training, the best 2,3,5 features
results_by_num_best_features = {}
best_2_features = ['engine-size', 'curb-weight']
results_by_num_best_features['best_2_features'] = knn_train_test(best_2_features, 'price', cars_features)

best_3_features = ['engine-size', 'curb-weight', 'city-mpg']
results_by_num_best_features['best_3_features'] = knn_train_test(best_3_features, 'price', cars_features)

best_5_features = ['engine-size', 'curb-weight', 'city-mpg', 'highway-mpg', 'width']
results_by_num_best_features['best_5_features'] = knn_train_test(best_5_features, 'price', cars_features)

results_by_num_best_features
```

```python
# visualize the results thru plot
for k,v in results_by_num_best_features.items():
    x = list(v.keys())
    y = list(v.values())
    plt.plot(x, y)
    plt.xlabel('k value')
    plt.ylabel('RMSE')
```

From the visualization, we can see that k-value of 2 to 5 will likely generate the best prediction of car price.
